```python
from google.colab import files

upload = files.upload()

# importar o pandas
import pandas as pd

# import do display (exibir objetos formatados )
from IPython.display import display

nome_arquivo = list (upload.keys())[0]

df = pd.read_csv (nome_arquivo)

print("dataframe completo:")

#df

display(df)
```

```python
#criar uma siries a partir de uma determinada coluna

serie_producao = df ['Producao']

print("\n Série de Produção")
display(serie_producao.head(10))
```

```python
#criar uma série calculada (producao - defeitos = peças boas)

serie_pecas_boas = df ['Producao'] - df ['Defeitos']

print("\n Série de Peças boas")
display(serie_pecas_boas)
```

```python
#adicionar essa serie criada ao dataframe

df ['Pecas_boas'] = serie_pecas_boas

print("dataframe atualizado:")
display(df)
```

```python
#  calcular a eficiencia (produção / tempo)

df ['Eficiencia(peca/min)'] = round(df ['Producao'] / df ['Tempo (min)'])

print("\ndataframe atualizado:")
display(df)
```

```python
#curiosidades que podem ser úteis

#listar os nomes das colunas

#print(df.columns.tolist())

#resumo descritivo do df
print (display(df.describe()))

#média dos valores da coluna
print ("média de produção: ")
df ['Producao'].mean()

#maior valor para coluna defeito
print ("\n maior valor de defeito: ")
df ['Defeitos'].max()

#soma dos valores de uma coluna
print ("\n soma dos valores de produção: ",df ['Producao'].sum())

```

```python
#geração de grafico com Matplotlib

import matplotlib.pyplot as plt

# produção total por dia- grafico de linha
# agrupar por dia

agrupado = df.groupby("Setor")
#print(agrupado)

#selecionar a coluna (serie) "producao"
selecionado = agrupado["Defeitos"]



display(selecionado.sum().plot(kind="line",marker="o", title="Produção por dia",ylabel=" N° Produção"))

plt.xticks(rotation=45)

```

```python
#geração de grafico com Matplotlib

import matplotlib.pyplot as plt

# produção total por dia- grafico de linha
# agrupar por dia

agrupado = df.groupby("Dia")
#print(agrupado)

#selecionar a coluna (serie) "producao"
selecionado = agrupado["Defeitos"]



display(selecionado.sum().plot(kind="bar", title="Produção por dia",ylabel=" N° Produção",color="y"))
```

```python
import matplotlib.pyplot as plt

#soma das pecas com defeitos

total_defeitos = df['Defeitos'].sum()

#soma das peças boas

total_pecas_boas = df['Pecas_boas'].sum()

plt.pie([total_defeitos,total_pecas_boas], labels=["Peças com defeitos", "Peças aprovada"], autopct="%1.1f%%", shadow= True, explode= [0.1,0.1])
plt.title("Qualidade da produção")
plt.legend(title = " peças")
plt.show()
```
