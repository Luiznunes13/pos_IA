```python
import pandas as pd

maps_df = pd.read_csv("maps_stats.csv")

print("dados carregados com sucesso")

for index, row in maps_df.iterrows():
    attack_win = float(row['attack_win_percent'].strip('%'))
    defense_win = float(row["defense_win_percent"].strip('%'))

    if attack_win > defense_win:
        print(f"No mapa {row['map_name']}, o ataque é mais forte")
    elif defense_win > attack_win:
        print(f"No mapa {row['map_name']}, a defesa é mais forte")
    else:
        print(f"No mapa {row['map_name']}, ataque e defesa são igualmente fortes")        
```
